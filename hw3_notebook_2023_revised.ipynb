Suppose we have a function $f(x)$ and we want to
calculate its derivative at a point $x$.  We can do that with pencil and
paper if we know the mathematical form of the function, or we can do it on
the computer by making use of the definition of the derivative:

\begin{equation*}
{ df\over dx} = \lim_{\delta\to0} {f(x+\delta)-f(x)\over\delta}.
\end{equation*}

On the computer we can't actually take the limit as $\delta$ goes to
zero, but we can get a reasonable approximation just by making $\delta$
small.

**a)** Write a program that defines a function _f(x)_ returning the
  value $x(x-1)$, then calculates the derivative of the function at the
  point $x=1$ using the formula above with $\delta=10^{-2}$.  


```python
#### PH 322 - HW #3, Q4.3a&b) - Dan Jang
### Credits to ChatGPT for math-code implementation help

## Constants
x = 1  # Point at which to calculate the derivative
delta = 10e-2

# Definition of the to-be-derivative function
def f(x):
    return x*(x - 1)  # Example function: f(x) = x^2

# The derivative calculation function
def derivative(f, x, delta):
    return (f(x + delta) - f(x - delta)) / (2 * delta)

def main():
    approx_derivative = derivative(f, x, delta)
    print("The approximate derivative of f(x) at x = " + str(x) + " & delta = " + str(delta) + " was found to be: " + str(approx_derivative))

    ## Output statements for Q4.3b.

    print("[For Q4.3b, #1] - The approximate derivative of f(x) at x = " + str(x) + " & delta = " + str(10e-4) + " was found to be: " + str(derivative(f, x, 10e-4)))
    print("[For Q4.3b, #2] - The approximate derivative of f(x) at x = " + str(x) + " & delta = " + str(10e-6) + " was found to be: " + str(derivative(f, x, 10e-6)))
    print("[For Q4.3b, #3] - The approximate derivative of f(x) at x = " + str(x) + " & delta = " + str(10e-8) + " was found to be: " + str(derivative(f, x, 10e-8)))
    print("[For Q4.3b, #4] - The approximate derivative of f(x) at x = " + str(x) + " & delta = " + str(10e-10) + " was found to be: " + str(derivative(f, x, 10e-10)))
    print("[For Q4.3b, #5] - The approximate derivative of f(x) at x = " + str(x) + " & delta = " + str(10e-12) + " was found to be: " + str(derivative(f, x, 10e-12)))
    print("[For Q4.3b, #6] - The approximate derivative of f(x) at x = " + str(x) + " & delta = " + str(10e-14) + " was found to be: " + str(derivative(f, x, 10e-14)))

if __name__ == "__main__":
    main();
```

Consider the integral

\begin{equation*}
E(x) = \int_0^x e^{-t^2} \>d t.
\end{equation*}

**a)** Write a program to calculate $E(x)$ for values of $x$ from 0 to 3 in
  steps of 0.1 using the **trapezoidal rule** and 20 slices. 

```python
#### PH 322 - HW #3, Q5.3a) - Dan Jang
### Credits to ChatGPT for math-code implementation help

## Libraries
import numpy as np

def f(t):
    return np.exp(-t**2)

def trapezoidal_integration(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    return h * (np.sum(y) - (y[0] + y[-1]) / 2)

def main():
    ## Constants
    n = 20
    a = 0
    b_values = np.arange(0, 3.1, 0.1)  # Values of x from 0 to 3 in steps of 0.1

    for b in b_values:
        E_x = trapezoidal_integration(f, a, b, n)
        print(f"Trapezoidal: E({b:.1f}) = {E_x:.6f}")

if __name__ == "__main__":
    main()
```

## Exercise 5.4

Our ability to resolve detail in astronomical observations is
limited by the diffraction of light in our telescopes.  Light from stars
can be treated effectively as coming from a point source at infinity.  When
such light, with wavelength $\lambda$, passes through the circular aperture
of a telescope (which we'll assume to have unit radius) and is focused by
the telescope in the focal plane, it produces not a single dot, but a
circular diffraction pattern consisting of central spot surrounded by a
series of concentric rings.  The intensity of the light in this diffraction
pattern is given by

\begin{equation*}
I(r) = \biggl( {J_1(kr)\over kr} \biggr)^2,
\end{equation*}

where $r$ is the distance in the focal plane from the center of the
diffraction pattern, $k=2\pi/\lambda$, and $J_1(x)$ is a Bessel function.
The Bessel functions $J_m(x)$ are given by

\begin{equation*}
J_m(x) = {1\over\pi} \int_0^\pi \cos(m\theta - x\sin\theta) \>d\theta,
\end{equation*}

where $m$ is a nonnegative integer and $x\ge0$.

**a)** Write a Python function _J(m,x)_ that calculates the value of
  $J_m(x)$ using Simpson's rule with $N=1000$ points.  Use your
  function in a program to make a plot, on a single graph, of the Bessel
  functions $J_0$, $J_1$, and $J_2$ as a function of $x$ from $x=0$ to
  $x=20$.

```python
#### PH 322 - HW #3, Q5.4a) - Dan Jang
### Credits to ChatGPT for math-code implementation & plotting help

## Libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def J(m, x):
    f = lambda t: np.cos(m * t - x * np.sin(t))
    return simpsons_rule_integration(f, 0, np.pi, 1000) / np.pi

# Define the Simpson's rule integration function
def simpsons_rule_integration(f, a, b, n):
    if n % 2 == 1:
        raise ValueError("The number of slices must be even for Simpson's rule.")
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    return h / 3 * (y[0] + y[-1] + 2 * np.sum(y[2:-1:2]) +
                    4 * np.sum(y[1:-1:2]))

def main():
    ## Constants
    lambda_val = 1.0
    k = 2 * np.pi / lambda_val
    r = 1.0

    # Define the function I(r)
    def I(r, k):
        return (J(1, k * r) / (k * r))**2


    # Create a range of values for x
    x_values = np.linspace(0, 20, 200)

    # Calculate the values of J_0, J_1, and J_2 for each value of x
    J_0_values = [J(0, x) for x in x_values]
    J_1_values = [J(1, x) for x in x_values]
    J_2_values = [J(2, x) for x in x_values]

    # Plot the results
    plt.plot(x_values, J_0_values, label="J_0")
    plt.plot(x_values, J_1_values, label="J_1")
    plt.plot(x_values, J_2_values, label="J_2")
    plt.xlabel("x")
    plt.ylabel("J_m(x)")
    plt.title("PH322, HW #3 - Q5.4a.) Bessel functions (J0, J1, & J2) for 0 <= x <= 20")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#### PH 322 - HW #3, Q5.4b) - Dan Jang
### Credits to ChatGPT for math-code implementation & plotting help

## Libraries
import numpy as np
import matplotlib.pyplot as plt

## Constants
wavelength = 500e-9  # 500 nm
k = 2 * np.pi / wavelength
lambda_val = 1.0
k = 2 * np.pi / lambda_val
r = 1.0


# Define the function to integrate
def J(m, x):
    f = lambda t: np.cos(m * t - x * np.sin(t))
    return simpsons_rule_integration(f, 0, np.pi, 1000) / np.pi


# Define the Simpson's rule integration function
def simpsons_rule_integration(f, a, b, n):
    if n % 2 == 1:
        raise ValueError(
            "The number of slices must be even for Simpson's rule.")
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    return h / 3 * (y[0] + y[-1] + 2 * np.sum(y[2:-1:2]) +
                    4 * np.sum(y[1:-1:2]))


# Define the function I(r)
def intensity(r, k):
    return (J(1, k * r) / (k * r))**2


def main():
    # Define the grid
    N = 1001
    r_max = 1e-6
    x = np.linspace(-r_max, r_max, N)
    y = np.linspace(-r_max, r_max, N)
    X, Y = np.meshgrid(x, y)
    r = np.sqrt(X**2 + Y**2)

    # Calculate the intensity
    intensity_value = (np.abs((2 * np.pi * J(1, k * r)) / (k * r)))**2

    # Plot the result
    fig, ax = plt.subplots()
    im = ax.imshow(intensity_value,
                   cmap='hot',
                   origin='lower',
                   extent=[-r_max, r_max, -r_max, r_max])
    ax.set_title('PH322, HW #3 - Q5.4b.) Density Plot of Diffraction Pattern')
    ax.set_xlabel('x (m)')
    ax.set_ylabel('y (m)')
    cbar = fig.colorbar(im)
    cbar.ax.set_ylabel('Intensity (W/m^2)')
    plt.show()


if __name__ == "__main__":
    main()
```
