### Fancier output formatting

Python gives you considerable control over the formatting of the output string. A detailed discussion of formatting can be found here [Link](https://docs.python.org/3/tutorial/inputoutput.html
"Fancier Output Formatting"). Here I provide a couple of examples.

Consider the following:

```python
a=1
b=3
c=a/b
print (c)
```

Let's say we want to format this to output only two decimals places. We can use a formatted string literal to specify the formatting by creating a string prefixed by 'f', as in:

```python
#print(f"The value of {a} divided by {b} is {c:.2f}.")
```

Here you can see I prefixed my string with 'f', then enclosed my variables in curly brackets which will output the values of the variables. In the third expression I follow my variable with :.2f. This tells Python to output my float to two decimal places. 

In the next example I want to format my output value using scientific notation.

```python
a=6.54e10
b=5
c=1.2e-4
d=a*b/c
print (d)
```

To do so, use the notation 'e' or 'E' in the format string:

```python
print(f"The value of d is {d:4.2e}.")
```

Here the 4 is the width or the number of characters in the coefficient or mantissa (including the decimal point), and 2 is the number of digits after the decimal to include. So '4.2e' means format the number using scientific notation where the coefficient has width 4 and 2 decimal places. 

We can also format the output using the string *format()* method. (Note, in Python a *method* is function that acts on an object, in this case the object is the string that we are outputting, and the *format()* method, well formats the string.)

Here is an example of the *format()* method:


```python
print("The value of d is {:4.2e}.".format(d))
```

Note I append the *format()* method to the end of the string with a dot, and specify the variable I wish to output. Within my string I use curly brackets to specify where in the string I want the variable to be placed and how I wanted it formatted.

The *format* method can contain any number of variables and I don't have to specify the format:

```python
print("The value of a, b and c are {}, {} and {}.".format(a,b,c))
```

# Importing functions
So far we have used a number of "built-in" Python functions, i.e. *print()*, *input()*, *float()*, and *int()*. These are functions in the core Python library that we can use "out of the box". Over the years developers have written other Python libraries that contain functions we would like to use. These libraries are called Python modules or packages. To use these modules we have to first download and install them in a location in our file system that Python knows about. Once we have the modules installed, we can use functions from these libraries using the **import** command. The modules we will use most often this term are:

* **numpy** - for numerical arrays and math functions that work with arrays
* **matplotlib** - for plotting and visualization
* **scipy** - for constants and other scientific computing tools
* **vpython** - for visualization and animation of 3D shapes
* **math** - contains lots of math functions for scalar variables
* **cmath** - contains complex math functions

Let's go through one example of how to import a function from a module.

Assume we want to find the cosine of the angle theta. Is there a cosine function in the standard Python library? Let's try it.


```python
# angle in radians
theta = 0
#cos(theta)
```

Oops! The interpretator tells us there is no cos() function define. 

We can see what built-in functions exist by looking at the directory of built-functions using the *dir()* command.

```python
dir(__builtins__)
```

If we scroll through the list we find our *float, input*, and *print* functions, but no *cos* function, nor any other math function. 

Where can we find a *cos* function? In the *math* module. To use it, we can import the math module into our current session using the *import* command:

```python
import math
```

We can now get a directory listing of the functions in the math module like we did above:

```python
dir(math)
```

Great, we see the *cos* function among others. Let's use it.

```python
#cos(theta)
```

What? Python says *cos* is not defined. What gives?

When we import an entire module, we have still have to tell Python where to look for the function by specifying the module name, a period, then the function name:

```python
math.cos(theta)
```

OK, now Python knows where to find the *cos* function and it works. 
You may think it's a pain to have to first refer to the module name. Is there another way?

Yes, one way is like this:

```python
from math import *
```

```python
cos(theta)
```

The asterisk is the wildcard character. It imports all the functions from the *math* module and allows us to use all the math functions without referring to the *math* module.

This seems like a win-win! 

Not necessarily. The same function name may be contained in more than one module. 
For example, there is also a *cos* function in the *numpy* module that works on arrays, whereas the *cos* function in the *math* module works only on scalars. If we were to import all functions from both modules using this method, we may not be able to keep them straight. In general it is best not to use " from <module> import * ".
    
There are a cpl of other option we can use however.  

```python
import math as m

m.cos(theta)
```

Here we use the "import <module> as " command. This allows us to use a shorthand abbreviation to refer to the module name when using the function, in this *m* for *math*.
    
We can also do this:

```python
from math import cos
cos (theta)
```

Here we import only the *cos* function from the *math* module. Doing this, we can use the function directly without referencing the module. 

We can use this syntax to import any number of functions from the same module. 

```python
from math import cos, sin, sqrt, exp
sin(theta)
```

Finally, some modules contain sub-modules which we import using both the module and submodule names:

```python
import matplotlib.pyplot as plt
plt.plot()
```

Here *pyplot* is a submodule of *matplotlib*, a popular module that contains plotting functions. We import the *pyplot* submodule as *plt* and use the *plot()* function to create an empty plot.
