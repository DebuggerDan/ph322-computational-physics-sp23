**c)** Technically a geosynchronous satellite is one that orbits the Earth once per sidereal
day, which is 23.93 hours, not 24 hours. Why is this? And how much difference
will it make to the altitude of the satellite?

Response:
A sidereal day for a geosynchronous satellite (GSO) is defined as 23 hours, 56 minutes, & 4 seconds, which is due to that the while Earth's rotation depends on the axis, it also revolves around the Sun.

Thus, the sidereal day represents one full rotation of the Earth relative to surrounding stars, whilst a 24-hour day is a 'solar day', simply the rotation around the Sun & not including relative axis rotation.

Thus, the 3 minutes & 56 seconds in difference does very slightly affect the altitude of the satellite, where we can modify our equation from Q2.2 as follows:

```python
#code for ex2.7

### PH 322 - HW #1 Q2.7.
# Credits to ChatGPT.

def catalan_number(n):
    return math.comb(2 * n, n) // (n + 1)

n = 0
max_value = 1000000000

while True:
    catalan = catalan_number(n)
    if catalan <= max_value:
        print(f"C{n} = {catalan}")
        n += 1
    else:
        break
```
